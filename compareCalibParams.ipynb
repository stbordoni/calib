```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
df_calib_SN04 = pd.read_csv('./calibDB/calib_biasV_SN04_v6_forcheck.csv',delimiter=',') 
df_calib_SN11 = pd.read_csv('./calibDB/calib_biasV_SN11_v2_forcheck.csv',delimiter=',') 
df_calib_SN15 = pd.read_csv('./calibDB/calib_biasV_SN15_v0_forcheck.csv',delimiter=',') 


df_calib_SN04.rename(columns={'Unnamed: 0':'Ch'},inplace=True)
df_calib_SN11.rename(columns={'Unnamed: 0':'Ch'},inplace=True)
df_calib_SN15.rename(columns={'Unnamed: 0':'Ch'},inplace=True)

```

```python
df_calib_SN04
```

```python
plt.figure(figsize=(10,6))

ax = plt.gca()

ax = df_calib_SN04.plot.scatter(x='Ch', y='Gain_Usig BiasV', color="Pink", label="Gain Bias Voltage SN04", ax=ax, marker = 'o', s=40 )
ax = df_calib_SN11.plot.scatter(x='Ch', y='Gain_Usig BiasV', color="Blue", label="Gain Bias Voltage SN11", ax=ax, marker = 'o', s=40 )
ax = df_calib_SN15.plot.scatter(x='Ch', y='Gain_Usig BiasV', color="Yellow", label="Gain Bias Voltage SN15", ax=ax, marker = 'o', s=40 )


plt.grid(b=True)
plt.title('Gain Bias Voltage')
plt.xlabel('Channel')
plt.ylabel('Bias Voltage Gain [V]')

plt.show()
```

```python
plt.figure(figsize=(10,6))

ax = plt.gca()

ax = df_calib_SN04.plot.scatter(x='Ch', y='Offset_Sig BiasV', color="Pink", label="Gain Bias Voltage SN04", ax=ax, marker = 'o', s=40 )
ax = df_calib_SN11.plot.scatter(x='Ch', y='Offset_Sig BiasV', color="Blue", label="Gain Bias Voltage SN11", ax=ax, marker = 'o', s=40 )
ax = df_calib_SN15.plot.scatter(x='Ch', y='Offset_Sig BiasV', color="Yellow", label="Gain Bias Voltage SN15", ax=ax, marker = 'o', s=40 )

plt.grid(b=True)
plt.title('Offset Bias Voltage')
plt.xlabel('Channel')
plt.ylabel('Bias Voltage Offset [V]')

plt.show()
```
